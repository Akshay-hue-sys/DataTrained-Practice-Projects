## OUTLIERS DETECTION AND REMOVAL:-

```python
plt.figure(figsize=(12,8),dpi=200)
sns.boxplot(data=df.iloc[:,1:],palette='colorblind')
plt.xticks(rotation=90);
```

There are some outliers present in the dataset shown by boxplot denoted by 'black diamonds'.

```python
#Outliers Removal:-
from scipy.stats import zscore
threshold=3
z=np.abs(zscore(df))
dfn=df[(z<3).all(axis=1)]
print(dfn.shape)
print(df.shape)
```

with the help of z-score standardisation, outliers have been removed.

```python
149/158*100
```

the data  defined as 'dfn' contains around 94% of the data after removing outliers.

```python
dfn.skew()
```

Skewness in the range of (-0.5,0.5) are normal.
The skewness can be seen as too much in some variables like generosity,family,health.
The extreme skewness needs to be adjusted as it can render our model prediction ineffective.

```python
#Skewness removal:-
for i in dfn.columns:
    if dfn[i].skew()>0.5:
        dfn[i]=np.cbrt(dfn[i])
    elif dfn[i].skew()<-0.5:
        dfn[i]=np.cbrt(dfn[i])
    else:
        pass
```

# MACHINE LEARNING PROCESS:-

```python
#seperating the target lable and features:-
X=dfn.drop('Happiness Score',axis=1) 
y=dfn['Happiness Score']
```

### Importing the necessary models for regression tasks and cross validation:-

```python
from sklearn.model_selection import cross_val_score,cross_validate
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
en = ElasticNet()
dtr = DecisionTreeRegressor(random_state=0)
knr = KNeighborsRegressor()
rfr = RandomForestRegressor(random_state=0)
ar = AdaBoostRegressor(random_state=0)
gbr= GradientBoostingRegressor(random_state=0)
sr = SVR()
list1 = [dtr,rfr,ar,gbr]
list2= [en,dtr,rfr,ar,gbr,sr]
```

### choosing the best random state:-

```python
min_squared_error = 0
max_random_state = 0
for i in range(1,200):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    model=DecisionTreeRegressor(random_state=i)
    model.fit(X_train,y_train)
    pred= model.predict(X_test)
    root_mean_squared_error=np.sqrt(mean_squared_error(y_test,pred))
    if root_mean_squared_error<min_squared_error:
        min_squared_error=root_mean_squared_error
        max_random_state=i
        
print('lowest error is: ',min_squared_error,' on random_state ',max_random_state)   
```

```python
#seperating the training data and test data:-
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### SCALING AND STANDARDIZATION:-

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
sX_train = s.fit_transform(X_train)
sX_test = s.transform(X_test)
```

Only training data has been both fit and transformed but the test data has only been transformed. '.fit' means calculating necessary statistics from the data and '.transform' only scales the data. test data should not be fit because that would lead to assumption of the prior knowledge of the test data and test data is the data, machine learning model need to predict fairly without having any knowledge of the test data.

```python
from sklearn.metrics import SCORERS
s=SCORERS
s.keys()
```

```python
for m in list2:
    m.fit(sX_train,y_train)
    preds = m.predict(sX_test)
    rmse= np.sqrt(mean_squared_error(y_test,preds))
    print(m)
    print('Root_mean_squared_error: ',rmse)
    print('\n')
    print('cross validation scores below:--  ',m)
    print('root_mean_squared_error: ',cross_val_score(m,X,y,cv=5,scoring='neg_root_mean_squared_error').mean())
    print('\n')
    
    
```

```python
def scores_(model):
    model.fit(sX_train,y_train)
    preds = model.predict(sX_test)
    rmse= np.sqrt(mean_squared_error(y_test,preds))
    
    scores_comparison = pd.DataFrame(list(zip([-rmse],
                               
                                [cross_val_score(model,X,y,cv=5,scoring='neg_root_mean_squared_error').mean()],
                               )),
                                     index=['Root_mean_squared_error'],
               columns=['model_scores','model_cross_validation_scores'])
    scores_comparison['difference_in_scores'] = scores_comparison.apply(lambda x: x['model_scores'] - 
                                                                        x['model_cross_validation_scores'], axis=1)
    return scores_comparison
```

rmse has been converted to -rmse for easy comparison,the higher the negative value the better the model is.

```python
dtr_ =scores_(dtr)
```

```python
en_=scores_(en)
```

```python
rfr_=scores_(rfr)
```

```python
ada_=scores_(ar)
```

```python
gbr_=scores_(gbr)
```

```python
sr_=scores_(sr)
```

```python
scoresfinal_ = pd.concat([en_,dtr_,rfr_,ada_,gbr_,sr_])
```

### GRID SEARCH CV:-

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
dtprr = {'max_features':["auto", "sqrt", "log2"],'criterion': ["mse", "friedman_mse", "mae"]}
```

```python

dtg=GridSearchCV(estimator=dtr,param_grid=dtprr)

print(dtg)
dtg.fit(sX_train,y_train)
y_pred=dtg.predict(sX_test)
print('Mean_absolute_error: ',mean_absolute_error(y_test,y_pred))
print('Mean_squared_error: ',mean_squared_error(y_test,y_pred))
print('Root_mean_squared_error: ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score: ',r2_score(y_test,y_pred))
print(dtg.best_params_)
```

```python
rfprr = {'n_estimators':[100,150,200,300,160],'max_features':['auto','log2','sqrt']}
gbprr = {'n_estimators':[150,200,300,250],'learning_rate':np.linspace(0.1,5,15),'max_depth':[5,6],
        'max_features':['auto','log2','sqrt']}


g1r = GridSearchCV(estimator=rfr,param_grid=rfprr)
g6r=GridSearchCV(gbr,param_grid=gbprr)


grs1 = [g1r,g6r]
for a in grs1:
    print(a)
    a.fit(sX_train,y_train)
    y_pred=a.predict(sX_test)
    print('Mean_absolute_error: ',mean_absolute_error(y_test,y_pred))
    print('Mean_squared_error: ',mean_squared_error(y_test,y_pred))
    print('Root_mean_squared_error: ',np.sqrt(mean_squared_error(y_test,y_pred)))
    print('R2_score: ',r2_score(y_test,y_pred))
    print(a.best_params_)
    print('\n')
```

The performance of the model gradient boosting regressor with grid search was better than random forest regressor.

### RANDOMISED SEARCH CV:-

```python
rsgbr= RandomizedSearchCV(estimator=gbr,param_distributions=rfprr,n_iter=15,scoring='neg_root_mean_squared_error')
```

```python
print(rsgbr)
rsgbr.fit(sX_train,y_train)
y_pred=rsgbr.predict(sX_test)
print('Mean_absolute_error: ',mean_absolute_error(y_test,y_pred))
print('Mean_squared_error: ',mean_squared_error(y_test,y_pred))
print('Root_mean_squared_error: ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score: ',r2_score(y_test,y_pred))
print(rsgbr.best_params_)
```

randomised search cv performs better than grid search cv for gradient boosting regressor as we can see from the comparison of the above metrics(root_mean_squared_error,mean_squared_error,mean_absolute_error and r2_score).
